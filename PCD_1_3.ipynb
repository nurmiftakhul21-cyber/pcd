```python
#----------NOMOR 3--------
from google.colab import files
uploaded = files.upload()

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('kucing.jpg', cv2.IMREAD_GRAYSCALE)

img_2bit = (img // 64) * 64
img_4bit = (img // 16) * 16
img_8bit = img

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Grayscale Asli")

plt.subplot(2,2,2)
plt.imshow(img_2bit, cmap='gray')
plt.title("2 bit (4 level)")

plt.subplot(2,2,3)
plt.imshow(img_4bit, cmap='gray')
plt.title("4 bit (16 level)")

plt.subplot(2,2,4)
plt.imshow(img_8bit, cmap='gray')
plt.title("8 bit (256 level)")

plt.show()
```
