```python
#----- lowpass filter -----
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Matriks dari soal (contoh dosen)
img = np.array([
    [6,4,2,1],
    [8,10,7,2],
    [8,10,7,2],
    [6,4,2,1]
], dtype=np.float32)

# Kernel Low Pass (jumlah = 1)
kernel = np.ones((3,3)) / 9

# Konvolusi
lowpass = cv2.filter2D(img, -1, kernel)

# Tampilkan
print("Low Pass Result:\n", lowpass)

plt.imshow(lowpass, cmap='gray')
plt.title("Low Pass")
plt.colorbar()
plt.show()
```

```python
#----- High pass filter -----
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.array([
    [6,4,2,1],
    [8,10,7,2],
    [8,10,7,2],
    [6,4,2,1]
], dtype=np.float32)

# Kernel High Pass
kernel = np.array([
    [-1,-1,-1],
    [-1, 8,-1],
    [-1,-1,-1]
])

# Konvolusi
highpass = cv2.filter2D(img, -1, kernel)

print("High Pass Result:\n", highpass)

plt.imshow(highpass, cmap='gray')
plt.title("High Pass")
plt.colorbar()
plt.show()
```

```python
#----- Band pass Filter -----
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.array([
    [6,4,2,1],
    [8,10,7,2],
    [8,10,7,2],
    [6,4,2,1]
], dtype=np.float32)

# Kernel Band Pass (contoh sharpen)
kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
])

# Konvolusi
bandpass = cv2.filter2D(img, -1, kernel)

print("Band Pass Result:\n", bandpass)

plt.imshow(bandpass, cmap='gray')
plt.title("Band Pass")
plt.colorbar()
plt.show()
```
