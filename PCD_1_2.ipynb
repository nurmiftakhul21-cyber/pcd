```python
from google.colab import files
uploaded = files.upload()

import cv2
import matplotlib.pyplot as plt

# membaca gambar
img = cv2.imread('kucing.jpg')

# cek apakah gambar terbaca
if img is None:
    print("Gambar tidak ditemukan, pastikan nama file benar")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img512 = cv2.resize(img,(512,512))
    img256 = cv2.resize(img,(256,256))
    img128 = cv2.resize(img,(128,128))
    img64 = cv2.resize(img,(64,64))

    plt.figure(figsize=(8,8))

    plt.subplot(2,2,1)
    plt.imshow(img512)
    plt.title("512x512")

    plt.subplot(2,2,2)
    plt.imshow(img256)
    plt.title("256x256")

    plt.subplot(2,2,3)
    plt.imshow(img128)
    plt.title("128x128")

    plt.subplot(2,2,4)
    plt.imshow(img64)
    plt.title("64x64")

    plt.show()
```
