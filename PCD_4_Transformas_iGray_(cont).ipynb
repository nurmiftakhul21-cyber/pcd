```python
#1----- Konversi RGB ke Grayscale---
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Before (RGB)")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
plt.title("After (Grayscale)")
plt.axis('off')

plt.show()
```

```python
#2-----Konversi Grayscale ke Biner-----
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Before (Grayscale)")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(binary, cmap='gray')
plt.title("After (Binary)")
plt.axis('off')

plt.show()
```

```python
#3-----Konversi Grayscale ke Mbit----
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

m = 4
levels = 2**m

gray_mbit = np.floor(gray / (256/levels)) * (256/levels)
gray_mbit = gray_mbit.astype(np.uint8)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Before (Grayscale)")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gray_mbit, cmap='gray')
plt.title("After (4-bit)")
plt.axis('off')

plt.show()
```

```python
#4-----image dengan brightness-----
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

brightness = cv2.convertScaleAbs(gray, alpha=1, beta=50)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Before (image)")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(brightness, cmap='gray')
plt.title("After (Brightness)")
plt.axis('off')

plt.show()
```

```python
#5-----Image Dengan Contrast-----
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

contrast = cv2.convertScaleAbs(gray, alpha=1.5, beta=0)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Before (Image)")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(contrast, cmap='gray')
plt.title("After (Contrast)")
plt.axis('off')

plt.show()
```

```python
#6----- Image Grayscale to Histogram-----
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title("Histogram")

plt.show()
```

```python
#7-----Image dengan Operasi Histogram Equalization-----
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()
img = cv2.imread(list(uploaded.keys())[0])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist_eq = cv2.equalizeHist(gray)

# Perbesar ukuran figure
plt.figure(figsize=(12,8))

# Tambahkan spacing biar tidak nabrak
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Before Image
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Before")
plt.axis('off')

# Histogram Before
plt.subplot(2,2,2)
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title("Histogram Before")

# After Image
plt.subplot(2,2,3)
plt.imshow(hist_eq, cmap='gray')
plt.title("After Equalization")
plt.axis('off')

# Histogram After
plt.subplot(2,2,4)
plt.hist(hist_eq.ravel(), bins=256, range=[0,256])
plt.title("Histogram After")

plt.show()
```
